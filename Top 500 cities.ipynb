# Top 500 India Cities.

EDA and Visualization.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cities=pd.read_csv(r"D:\python\cities.csv")
```

```python
cities.head().T
```

```python
cities.info()
```

```python
cities.describe()
```

```python
cities.isnull().sum()
```

```python
cities=cities.drop_duplicates(keep="first")
```

```python
cities.shape
```

```python
cities.state_name.value_counts()
```

```python
f,ax=plt.subplots(figsize=(16,10)) 
sns.countplot("state_name",data=cities)
plt.xticks(rotation=90,fontsize=15)
plt.xlabel("State_Name",fontsize=20)
plt.ylabel("Cities",fontsize=20)
```

```python
states_with_more_cities=cities.state_name.value_counts(sort=True,ascending=False)
topstates=states_with_more_cities.head(10)
bottomstates=states_with_more_cities.tail(10)
print(topstates,bottomstates)
```

```python
cities_state=cities.groupby("state_name").sum()
cities_state
```

```python
cities_num=cities.select_dtypes(include=["int64","float64"])
cities_num
```

```python
cities_cat=cities.select_dtypes(include=["object"])
cities_cat
```

```python
corr=cities_num.corr()
plt.subplots(figsize=(18,15))
sns.heatmap(corr,annot=True,square=True,linewidths=0.4,linecolor="r")
```

```python
data=cities.sort_values("population_total",ascending=False)
top10=data.head(10)
bottom10=data.tail(10)
```

```python
plt.bar(top10['name_of_city'],top10['population_male'],label='Male Population',color='blue',alpha=0.4)
plt.bar(top10['name_of_city'],top10['population_female'],label='Female Population',color='red',alpha=0.4)
plt.bar(top10["name_of_city"],top10["literates_male"],label="Male Literates",color="orange")
plt.bar(top10["name_of_city"],top10["literates_female"],label="Female Literates",color="m")
plt.xticks(rotation=90,fontsize=15)
plt.xlabel('City Names',fontsize=15)
plt.title('Literacy rate in top10 cities by population',fontsize=20)
plt.legend(loc=0,fontsize=15)
plt.show()
```

```python
plt.bar(bottom10["name_of_city"],bottom10["population_male"],label="Male population")
plt.bar(bottom10["name_of_city"],bottom10["population_female"],label="Female population")
plt.bar(bottom10["name_of_city"],bottom10["literates_male"],label="Male Literates")
plt.bar(bottom10["name_of_city"],bottom10["literates_female"],label="Female Literates")
plt.xticks(rotation=90,fontsize=15)
plt.xlabel("City Names",fontsize=15)
plt.title("Literacy rate in Bottom 10 cities by population")
plt.legend(fontsize=10,loc=8)
plt.show()
```

```python
plt.bar(top10["name_of_city"],top10["population_total"],label="Total Population",color="b")
plt.bar(top10["name_of_city"],top10["literates_total"],label="Total Literates",color="red")
plt.bar(top10["name_of_city"],top10["total_graduates"],label="Total Graduates",color="orange")
plt.xticks(rotation=90,fontsize=15)
plt.xlabel("City Names",fontsize=15)
plt.title("Population vs Graduates in top10 cities")
plt.legend(fontsize=10)
plt.show()
```

```python
plt.bar(bottom10["name_of_city"],bottom10["population_total"],label="Total Population",color="green",alpha=0.4)
plt.bar(bottom10["name_of_city"],bottom10["literates_total"],label="Total Literates",color="b",alpha=0.4)
plt.bar(bottom10["name_of_city"],bottom10["total_graduates"],label="Total Graduates",color="black",alpha=0.4)
plt.xticks(rotation=90,fontsize=15)
plt.xlabel("City Names",fontsize=15)
plt.title("Population vs Graduates in bottom 10 cities")
plt.legend(fontsize=10)
plt.show()
```

```python
f,ax=plt.subplots(figsize=(18,15))
sns.boxplot(x="state_name",y="total_graduates",data=cities)
plt.xticks(rotation=90,fontsize=10)
```

```python
f, ax = plt.subplots(figsize=(8, 12))
sns.barplot(x="population_total",y="state_name",data=cities,color="b",label="Population")
sns.barplot(x="literates_total",y="state_name",data=cities,color="r",label="Literates")
sns.barplot(x="total_graduates",y="state_name",data=cities,color="black",label="Graduates")
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
plt.ylabel("State Names")
plt.legend(fontsize=15)
plt.show()
```
